```python
import gpxpy
import gpxpy.gpx

# gpx_files = ["PT22 Dag 1.gpx","2020 dag 2 11-01-20.gpx","PT22 Dag 3.gpx","PT22 Dag 4 (6 daagse) 1e.gpx","PT22 dag 5 2e.gpx","PT22 Dag 6 2e.gpx","Routevoorstel PT22 dag 4(4daagse)1e.gpx"]
# days = ["dag 1","dag 2","dag 3","dag 4 (6 daagse)","dag 5","dag 6","dag 4 (4 daagse)"]

gpx_files = ["HeH22 dag 1 Palogne Hamoir.gpx","HeH22 dag 2 Hamoir Rivage afkorting.gpx","HeH22 dag 2 Hamoir Rivage.gpx","HeH22 dag 3 Rivage Aywaille.gpx","HeH22 dag 4 Aywaille Remouchamps.gpx"]
days = ["dag 1","dag 2 (afkorting)","dag 2","dag 3","dag 4"]

map_boundaries = [0]*4 #[min_lon, max_lon, min_lat, max_lat]
route_coordinates = {}
for index,gpx_file in enumerate(gpx_files):
#     gpx_file = open("PT22GPX/"+gpx_file, 'r')
    gpx_file = open(gpx_file, 'r')
    gpx = gpxpy.parse(gpx_file)
    coordinates_list = []
    for track in gpx.tracks:
        for segment in track.segments:
            for point in segment.points:
                coordinates_list.append(point.longitude)
                coordinates_list.append(point.latitude)
                if index==0:
                    map_boundaries = [point.longitude,point.longitude,point.latitude,point.latitude]
                if map_boundaries[0]>point.longitude:
                    map_boundaries[0]=point.longitude
                if map_boundaries[1]<point.longitude:
                    map_boundaries[1]=point.longitude
                if map_boundaries[2]>point.latitude:
                    map_boundaries[2]=point.latitude
                if map_boundaries[3]<point.latitude:
                    map_boundaries[3]=point.latitude      
                
    route_coordinates[days[index]] = coordinates_list

print(route_coordinates)
print(map_boundaries)
map_boundaries[0] = map_boundaries[0]-0.003
map_boundaries[1] = map_boundaries[1]+0.003
map_boundaries[2] = map_boundaries[2]-0.003
map_boundaries[3] = map_boundaries[3]+0.003
print(map_boundaries)
```

```python
import sqlite3

filename = "/Users/ernestjanssen/Documents/paastocht/APP/map_data_2022.mbtiles"
db_mbtiles = sqlite3.connect(filename)
c_mbtiles = db_mbtiles.cursor()
c_mbtiles.execute("CREATE TABLE IF NOT EXISTS `route_coordinates` (`day` varchar(20) NOT NULL, `coordinate_string` blob NOT NULL)")

for day in days:
    sql = "INSERT INTO route_coordinates (day,coordinate_string) VALUES (?,?)"
    val = (day,bytes(str(route_coordinates[day]),'utf-8'))
    c_mbtiles.execute(sql,val)
    
c_mbtiles.close()
db_mbtiles.commit()
db_mbtiles.close()

```
